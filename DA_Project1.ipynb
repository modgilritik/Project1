```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore", category=FutureWarning)

# Load the dataset into a DataFrame
df_15 = pd.read_csv(r'C:\Users\d7sha\OneDrive\Desktop\Assignments\DataAnalytics\2015_KSI.csv')
df_16 = pd.read_csv(r'C:\Users\d7sha\OneDrive\Desktop\Assignments\DataAnalytics\2016_KSI.csv')
df_17 = pd.read_csv(r'C:\Users\d7sha\OneDrive\Desktop\Assignments\DataAnalytics\2017_KSI.csv')
df_18 = pd.read_csv(r'C:\Users\d7sha\OneDrive\Desktop\Assignments\DataAnalytics\2018_KSI.csv')
```

```python
#Merge all data
df=pd.concat([df_15,df_16,df_17,df_18])
allData = df[['YEAR', 'VEHICLES_IN_STREET', 'DISTRICT', 'NEIGHBOURHOOD','ACCLASS']]
allData
```

```python
# Group the data by 'Year' and 'District' and calculate the total vehicles in the street
vehicles_per_district = allData.groupby(['YEAR', 'DISTRICT'])['VEHICLES_IN_STREET'].sum()

# Print the result
vehicles_per_district
```

```python
# Group the data by 'Neighbourhood' and calculate the average number of vehicles in the street
average_vehicles_per_neighbourhood = allData.groupby('NEIGHBOURHOOD')['VEHICLES_IN_STREET'].mean()

# Print the result
average_vehicles_per_neighbourhood
```
